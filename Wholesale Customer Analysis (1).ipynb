Descriptive measure of variablility describe the variability or spread in data set .The descriptive measures of variability are range ,IQR,Variance,standard deviation, and coefficient of variance.Here we will use coefficient of variance.

Coefficient of variance is statistical measure of dispersion of data points in a series around the mean .
In simple words ,its a ratio of standarad deviation to the mean.
We are using CV here as this is the most usefull statistical measure to compare the variability from one series to another ,even their means are different from one another.
It is best suited to our data set where the means are different.


```python
from scipy.stats import variation
```

```python
mydata.columns
```

```python
cols=['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicatessen']
```

```python
for i in cols:
  print('The Coefficient of variation for {} is {}'.format(i,round(variation(mydata[i]),3)))
```

All the numerical variable are stored in list named as cols .
taking i as iterating variable in this list of cols we are find the coefficient of variation value for each variable in data set.
two curly brackets are used in one each varible will comes iteratively ad second braces will give us the CV value for that variable.

Inference :
From the above ressults ,we found that Delicatessen shows the most inconsistent behaviour(High CV values=1.847) and Fresh shows the least inconsistent behaviour (lowest CV values=1.053)

Outlier Check in data:
In statsistics ,an outlier is an point that is distant from the other observations in data .
Boxplot is visual method used to detect the outliers in data .



```python
mydata.columns
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
fig.set_size_inches(18,10)
a=sns.boxplot(data=mydata,x='Fresh',orient="v",ax=axes[0][0])
a.set_title("Fresh Boxplot",fontsize=15)
b=sns.boxplot(data=mydata,x='Milk',orient="v",ax=axes[0][1])
b.set_title("Milk Boxplot",fontsize=15)
c=sns.boxplot(data=mydata,x='Grocery',orient="v",ax=axes[0][2])
c.set_title("Grocery Boxplot",fontsize=15)
d=sns.boxplot(data=mydata,x='Frozen',orient="v",ax=axes[1][0])
d.set_title("Frozen Boxplot",fontsize=15)
e=sns.boxplot(data=mydata,x='Detergents_Paper',orient="v",ax=axes[1][1])
e.set_title("Detergents_Paper Boxplot",fontsize=15)
f=sns.boxplot(data=mydata,x='Delicatessen',orient="v",ax=axes[1][2])
f.set_title("Delicatessen Boxplot",fontsize=15)
plt.show()
```

From the boxplots,we found that all the variables contains outliers.
